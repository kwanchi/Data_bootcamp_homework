```python
# Dependencies
import pandas as pd
import os
```

```python
# load CSV
path = os.path.join("Resources/2016-FCC-New-Coders-Survey-Data.csv")
coders_df = pd.read_csv(path)
coders_df.head()
```

```python
# Read with pandas
coders_df.head()
```

```python
# Inspect all columns
coders_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)

coders_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
converted_df = coders_df.replace([0.0, 1.0,], ["No","Yes"])
# Hint use the df.replace function
converted_df
```

```python
# Calculate total number of respondents in survey
len(converted_df)
# consider using the len() function
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_df = converted_df.loc[converted_df["AttendedBootcamp"] == "Yes"]
attended_df.tail()
# Filter using df.loc on the AttendedBootcamp column
```

```python
# Calculate average age of attendees
attended_df['Age']
# Pull out the `Age` series and take the average of it
attended_df['Age'].mean()
```

```python
# Calculate how many people attended a bootcamp
len(attended_df)
# Count the values of the AttendedBootcamp column
x = attended_df["AttendedBootcamp"].value_counts()
pd.DataFrame(x)
```

```python
# Calculate how many attendees hold degrees
y = attended_df["SchoolDegree"].value_counts()
# It is easier to do this with `value_counts() count values for the SchoolDegree column
y
```

```python
# Count the number of records where the person is a degree holder
deg_df = attended_df.replace(["bachelor's degree","Ph.D.", "master's degree (non-professional)","professional degree (MBA, MD, JD, etc.)","associate's degree"], ["Yes","Yes","Yes","Yes","Yes"])
deg_df.head()
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column
```

```python
b_df = deg_df.loc[deg_df['SchoolDegree'] == 'Yes']
b_df['SchoolDegree'].count()
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
attended_df['Gender'].value_counts()
#len(attended_df['Gender'])
```

```python
# Calculate percentage of respondents who attended a bootcamp
bootcamp_avg = len(attended_df) / len(coders_df) 
bootcamp_avg
```

```python
# Calculate percentage of respondents belonging to each gender
percentages = attended_df['Gender'].value_counts('female')
#len(attended_df['Gender'])
percentages
```

```python
male = percentages[0]
female = percentages[1]
queer = percentages[2]
trans = percentages[3]
agender = percentages[4]
non = queer + trans + agender
non
```

```python
# Calculate percentage of attendees with a college degree
percent_w_degrees = b_df['SchoolDegree'].count() / len(attended_df)
percent_w_degrees 
```

```python
# Calculate average post-bootcamp salary
salary = attended_df['BootcampPostSalary'].mean()
salary
```

```python
# Create a new table consolidating above calculations
final_bootcamp = pd.DataFrame({
    "Total Surveyed":[len(converted_df)], "Total Bootcamp Attendees":[len(attended_df)], "% Attended Bootcamp":[bootcamp_avg], "Average Age":[attended_df['Age'].mean()],"Has a Degree":[b_df['SchoolDegree'].count()], "% Male": [male],
"% Female":[female], "% Non Gender Specific": [non], "Average Post Bootcamp Salary": [salary] })

final_bootcamp
```

```python
# Improve formatting before outputting spreadsheet
final_formatted = final_bootcamp.style.format({"Total Surveyed": "{:2,}", "% Attended Bootcamp": "{:,.2%}" , "Average Age":"{:.0f}", "% Male":"{:,.2%}","% Female":"{:,.2%}", "% Non Gender Specific":"{:,.2%}", "Average Post Bootcamp Salary":"{:,.2f}"})
final_formatted
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes
final_formatted.to_excel('data_cleanup_output_KW.xlsx', index = False)
```
