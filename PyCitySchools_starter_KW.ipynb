# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset (consider using a left join)
merged_df = pd.merge(student_data, school_data, on="school_name", how = "left")
merged_df
```

```python
# Calculate the passing scores by creating a filtered data frame
reading = merged_df.loc[merged_df['reading_score'] >= 70]
school_reading = reading.groupby('school_name')
read_pass = school_reading.count()['reading_score']
reading_pass_pct = read_pass / student_count
reading_pass_pct
```

```python
# Calculate the passing scores by creating a filtered data frame
math = merged_df.loc[merged_df['math_score'] >= 70]
school_math = math.groupby('school_name')
math_pass = school_math.count()['math_score']
math_pass_pct = math_pass / student_count
math_pass_pct
total_pass_pct = (reading_pass_pct + math_pass_pct)/2
total_pass_pct
```

```python
# Convert to data frame
pass_df = pd.DataFrame({"Passing Percentages":total_pass_pct})

# Minor data munging
pass_df = pass_df["Passing Percentages"].map("{:.2%}".format)
# Display the data frame
pass_df
```
