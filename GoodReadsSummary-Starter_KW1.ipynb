```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Path and filename for the file to Load
books_path = os.path.join('Resources','books_clean.csv')
books_df = pd.read_csv(books_path, encoding='utf-8')
books_df.head()

# Read the modified GoodReads csv and store into Pandas DataFrame

```

```python
# Calculate the number of unique authors in the DataFrame
authors = books_df["Authors"].value_counts()
authors
authors_df = pd.DataFrame(authors)
authors_df

```

```python
authors_df.count()
```

```python
# Calculate the earliest/latest year a book was published
books_df['Publication Year'].min()
```

```python
# Calculate the total reviews column that contains the sum of reviews (total - all authors)
reviews_df = books_df[['One Star Reviews','Two Star Reviews','Three Star Reviews','Four Star Reviews','Five Star Reviews']]
reviews_df
# Hint: use the pandas' sum(axis='columns') to get the sum across the columns 
```

```python
# Create a `total_reviews` variable that sums up the values from your new `Total Reviews` column in the DataFrame
reviews_df["Total Reviews"] = reviews_df.sum(axis='columns')
reviews_df["Total Reviews"].sum()
```
